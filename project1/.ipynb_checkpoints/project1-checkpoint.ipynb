```python
#############################
#### With CV 
#############################

def best_degree_selection(degrees, k_fold, model, seed = 1):
    
    # Split data in k fold
    k_indices = build_k_indices(y_pre, k_fold, seed)
    
    # For each degree, store :
    best_param = []
    best_rmses = []
    best_accuracy = []
    
    model2 = model.copy()
    
    # Vary degree
    for i, degree in enumerate(degrees):
        
        print("Progress : ", i/len(degrees)*100 , " % ")
        
        store_rmse_te = []

        #### Ridge regression #####
        if model["model"] == 'ridge':
            
            params = model["param"]
            for p in params:
                
                rmse_te_tmp = []
                accuracy_tmp = []
                model2["param"] = p
                
                for k in range(k_fold):
                    _, rmse_te, _, accuracy = cross_validation(y_pre, tX_post, k_indices, k, model2, degree)
                    rmse_te_tmp.append(rmse_te)
                    accuracy_tmp.append(accuracy)
                store_rmse_te.append(np.mean(rmse_te_tmp))
                best_accuracy.append(np.mean(accuracy_tmp))
        
            ind_param_opt = np.argmin(store_rmse_te)
            best_param.append(gammas[ind_param_opt])
            best_rmses.append(store_rmse_te[ind_param_opt])
            
        elif model["model"] != 'ridge':
            
            #### GD or SGD #######
            if model["model"] == 'GD' or model["model"] == 'SGD':
                # weights
                model2["initial_w"] = np.zeros(int(tX.shape[1] * degree + 1))
            
            ##### LS ######
            rmse_te_tmp = []
            accuracy_tmp = []
            for k in range(k_fold):
                _, rmse_te, _, accuracy = cross_validation(y, tX, k_indices, k, model2, degree)
                rmse_te_tmp.append(rmse_te)
                accuracy_tmp.append(accuracy)
                
            best_rmses.append(np.mean(rmse_te_tmp))
            best_accuracy.append(np.mean(accuracy_tmp))

        
    ind_best_degree =  np.argmin(best_rmses)

        
    return degrees[ind_best_degree], best_accuracy, best_rmses
```

```python
################
# CV GD
###############

def CV_GD(degrees, gammas, max_iters, k_fold, seed = 1):
    
    # Split data in k fold
    k_indices = build_k_indices(y_pre, k_fold, seed)

    # For each degree, store :
    best_param = []
    best_rmses = []
    best_accuracy = []
    
    parameters_GD = {
        "model" : "GD",
        "param": [0, max_iters],
    }

    # Vary degree
    for i, degree in enumerate(degrees):

        print("Progress : ", i/len(degrees)*100 , " % ")
        
        parameters_GD["initial_w"] = np.zeros(int(tX_post.shape[1] * degree + 1))
        
        for gamma in gammas:
            
            parameters_GD["param"][0] = gamma
            print(gamma)
            
            # Contains the rmse/accuracy for 1 degree and all gammas
            rmse_gamma = []
            accuracy_gamma = []
            
            # Contains rmse or accuracy for each k_fold
            rmse_te_tmp = []
            accuracy_tmp = []
            
            for k in range(k_fold):
                _, rmse_te, _, accuracy = cross_validation(y_pre, tX_post, k_indices, k, parameters_GD, degree)
                rmse_te_tmp.append(rmse_te)
                accuracy_tmp.append(accuracy)

            rmse_gamma.append(np.mean(rmse_te_tmp))
            accuracy_gamma.append(np.mean(accuracy_tmp))
        
        idx_best_gamma =  np.argmin(rmse_gamma)
        best_param.append(gammas[idx_best_gamma])
        best_rmses.append(rmse_gamma[idx_best_gamma])
        best_accuracy.append(accuracy_gamma[idx_best_gamma])


    idx_best_degree =  np.argmin(best_rmses)


    return idx_best_degree, best_param, best_accuracy, best_rmses
```

```python
################
# CV LS
###############

def CV_LS(degrees, k_fold, seed = 1):
    
    # Split data in k fold
    k_indices = build_k_indices(y_pre, k_fold, seed)

    # For each degree, store :
    best_rmses = []
    best_accuracy = []
    
    parameters_LS = {
        "model" : "LS",
    }

    # Vary degree
    for i, degree in enumerate(degrees):

        print("Progress : ", i/len(degrees)*100 , " % ")
        

        # Contains rmse or accuracy for each k_fold
        rmse_te_tmp = []
        accuracy_tmp = []

        for k in range(k_fold):
            _, rmse_te, _, accuracy = cross_validation(y_pre, tX_post, k_indices, k, parameters_LS, degree)
            rmse_te_tmp.append(rmse_te)
            accuracy_tmp.append(accuracy)

        best_rmses.append(np.mean(rmse_te_tmp))
        best_accuracy.append(np.mean(accuracy_tmp))


    idx_best_degree =  np.argmin(best_rmses)


    return idx_best_degree, best_accuracy, best_rmses
```

```python
################
# CV Ridge
###############

def CV_Ridge(degrees, lambdas, k_fold, seed = 1):
    
    # Split data in k fold
    k_indices = build_k_indices(y_pre, k_fold, seed)

    # For each degree, store :
    best_param = []
    best_rmses = []
    best_accuracy = []
    
    parameters_ridge = {
      "model": "ridge",
      "param": lambdas,
    }

    # Vary degree
    for i, degree in enumerate(degrees):

        print("Progress : ", i/len(degrees)*100 , " % ")
        
        
        for lambda_ in lambdas:
            
            parameters_ridge["param"] = lambda_
            print(lambda_)
            
            # Contains the rmse/accuracy for 1 degree and all gammas
            rmse_lambda = []
            accuracy_lambda = []
            
            # Contains rmse or accuracy for each k_fold
            rmse_te_tmp = []
            accuracy_tmp = []
            
            for k in range(k_fold):
                _, rmse_te, _, accuracy = cross_validation(y_pre, tX_post, k_indices, k, parameters_ridge, degree)
                rmse_te_tmp.append(rmse_te)
                accuracy_tmp.append(accuracy)

            rmse_lambda.append(np.mean(rmse_te_tmp))
            accuracy_lambda.append(np.mean(accuracy_tmp))
        
        idx_best_lambda =  np.argmin(rmse_lambda)
        best_param.append(lambdas[idx_best_lambda])
        best_rmses.append(rmse_lambda[idx_best_lambda])
        best_accuracy.append(accuracy_lambda[idx_best_lambda])


    idx_best_degree =  np.argmin(best_rmses)


    return idx_best_degree, best_param, best_accuracy, best_rmses

```

```python
degrees = np.linspace(1, 20, 20)
lambdas = np.logspace(-4, 0, 10)
k_fold = 5

idx_best_degree_ridge, best_accuracy_ridge, best_rmses_ridge = CV_Ridge(degrees, lambdas, k_fold, seed = 1)
```

```python
print("The best degree is :", degrees[idx_best_degree_ridge])
print("The best gamma is :", lambdas[idx_best_degree_ridge])

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
stop = 20
ax1.plot(degrees[:stop], best_rmses_ridge[:stop], 'g-')
ax2.plot(degrees[:stop], best_accuracy_ridge[:stop], 'b-')

ax1.set_xlabel('Degree')
ax1.set_ylabel('RMSE', color='g')
ax2.set_ylabel('Accuracy in %', color='b')

plt.show()
```
